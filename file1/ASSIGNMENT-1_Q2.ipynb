# 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame

```python
# import required libraries
```

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page=requests.get('https://presidentofindia.nic.in/former-presidents')
```

```python
page
```

```python
soup=bs(page.content)
```

```python
soup
```

```python
# Scraping Name of formar Presidents of India
head=['h3']
name=[]

for i in soup.find_all(head):
    name.append(i.text)
```

```python
name
```

```python
head1=['h5']
termofoffice=[]

for i in soup.find_all(head1):
    termofoffice.append(i.text)
```

```python
termofoffice
```

```python
#Printing length

print (len(name), len(termofoffice))
```

```python
import pandas as pd
```

```python
#Data Frame

df=pd.DataFrame({'Name Of President Of India': name, 'Term of Office': termofoffice})
```
