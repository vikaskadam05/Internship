# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/t20i')
```

```python
page
```

```python
# Page Content
soup=bs(page.content)
```

```python
soup
```

```python
# Top 10 Teams
N=10
team=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text)
topteam=team[:N]
```

```python
topteam
```

```python
# Matches Stats
N=20
matches=[]

for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)
totalmatches=matches[:N]    
```

```python
totalmatches
```

```python
tmatch=totalmatches[0:20:2]
```

```python
tmatch
```

```python
tpoints=totalmatches[1:20:2]
```

```python
tpoints
```

```python
# Rating
N=10
rating=[]

for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
toprating=rating[:N]
```

```python
toprating
```

```python
print(len(topteam), len(totalmatches), len(toprating))
```

```python
df=pd.DataFrame({'Team':topteam, 'Total Matches':tmatch,'Total Points':tpoints,'Rating':toprating})
```

```python
df
```
