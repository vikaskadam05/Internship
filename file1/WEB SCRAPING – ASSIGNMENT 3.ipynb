# 7. Write a python program to scrape the details for all billionaires from www.forbes.com. 
Details to be scrapped: 
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
#Opening the map on forbes.com page on automated Chrome browser

driver=webdriver.Chrome()
driver.get("https://www.forbes.com/billionaires/")
```

```python
rank1=[]
name=[]
net_worth=[]
age=[]
citizenship=[]
source=[]
```

```python
#Scraping Rank, Name, Net worth, Age, Citizenship, Source, Industry.

start=0
end=13

for page in range (start,end):
    rank1_tag=driver.find_elements(By.XPATH,'//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]')
    for i in rank1_tag:
        rank1.append(i.text)
    
    name_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][2]')
    for i in name_tag:
        name.append(i.text)

    net_worth_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][3]')
    for i in net_worth_tag:
        net_worth.append(i.text)
    
    age_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][4]')
    for i in age_tag:
        age.append(i.text)
    
    citizenship_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][5]')
    for i in citizenship_tag:
        citizenship.append(i.text)
    
    source_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][6]')
    for i in source_tag:
        source.append(i.text)
    
           
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]/div/span[1]')
    next_button.click()
    time.sleep(5)
```

```python
industry_2=[]
start=0
end=13
for page in range (start,end):
    industry_2_tag=driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][7]')
    for i in industry_2_tag:
        industry_2.append(i.text)
    
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]/div/span[1]')
    next_button.click()
    time.sleep(5)
```

```python
print(len(rank1),len(name),len(net_worth),len(source),len(age),len(industry_2))
```

```python
#Searching London location on www.hostelworld.com

hotel=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button')
hotel.send_keys('London')

search_hotel=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button')
search_hotel.click()

search_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]')
search_button.click()
```

```python
#fetching url to open hotel

hotel_url=[]
start=0
end=3

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@data-v-547b8f9f]')
    for i in url:
        hotel_url.append(i.get_attribute('href'))
```

```python
len(hotel_url)
```

```python
#fetching hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

name=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        name_tag=driver.find_element(By.XPATH,'//h1[@data-v-e9561270]/div')
        name.append(name_tag.text)
    except NoSuchElementException:
        name.append('-')
```

```python
rating=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        rating_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[1]')
        rating.append(rating_tag.text)
    except NoSuchElementException:
        rating.append('-')
```

```python
totalreview=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        totalreview_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[1]/div[1]/div[2]/div[2]')
        totalreview.append(totalreview_tag.text)
    except NoSuchElementException:
        totalreview.append('-')
```

```python
overallreview=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        overallreview_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[11]/div[1]/ul')
        overallreview.append(overallreview_tag.text.replace('\n','- '))
    except NoSuchElementException:
        overallreview.append('-')
```

```python
privatesprice=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        privatesprice_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div/section/div/div/div[2]/div[1]/div[2]')
        privatesprice.append(privatesprice_tag.text.replace('\n',' '))
    except NoSuchElementException:
        privatesprice.append('-')
```

```python
dormprice=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        dormprice_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div/section/div/div/div[2]/div[2]/div[2]')
        dormprice.append(dormprice_tag.text.replace('\n',' '))
    except NoSuchElementException:
        dormprice.append('-')
```

```python
facilities=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        facilities_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div')
        facilities.append(facilities_tag.text.replace('\n',' '))
    except NoSuchElementException:
        facilities.append('-')
```

```python
description=[]

for i in hotel_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        description_tag=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[8]/div/div[2]/div[3]/div/div[1]')
        description.append(description_tag.text.replace('\n',' '))
    except NoSuchElementException:
        description.append('-')
```

```python
print(len(name), len(rating), len(totalreview), len(overallreview), len(privatesprice), len(dormprice), len(facilities), len(description))
```
