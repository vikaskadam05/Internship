B) Data type

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
#Opening the billboard.com page on automated Chrome browser

driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
delay=10
try:
    element=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a')
    element.click()
    print(element)
except StaleElementReferenceException as e:
    print("Exception Raised: ",e)
    print("Refershing the page!!")
    driver.get("https://archive.ics.uci.edu/")
    WebDriverWait(driver,delay).until(EC.presence_of_all_elements_located((By.CLASS_NAME,'dropdown-content')))
    element=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/header/nav/ul/li[1]/a')
    element.click()
    print(element)
```

```python
dn=[]
dt=[]
ts=[]
at=[]
ni=[]
na=[]
yr=[]
```

```python
start=0
end=10

for page in range (start,end):
    xpd_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[2]')
    xpd_button.click()
    xpd_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]/div[2]/span[1]')
    xpd_button.click()
    dn_tag=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
    for i in dn_tag:
        dn.append(i.text)
    dt_tag=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][2]')
    for i in dt_tag:
        dt.append(i.text)
    ts_tag=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][1]')
    for i in ts_tag:
        ts.append(i.text)
    at_tag=driver.find_elements(By.XPATH,'//td[/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[2]/div/table/tbody/tr][2]')
    for i in at_tag:
        at.append(i.text)
    ni_tag=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][3]')
    for i in ni_tag:
        ni.append(i.text)
    na_tag=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][4]/span')
    for i in na_tag:
        na.append(i.text)
    yr_tag=driver.find_elements(By.XPATH,'//td[/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[2]/div/table/tbody/tr][3]')
    for i in yr_tag:
        yr.append(i.text)
    
           
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    time.sleep(3)
```
