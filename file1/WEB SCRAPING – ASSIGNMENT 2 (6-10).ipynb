# Q6: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 3 attributes of each sneaker:
1. Brand
2. Product Description
3. Price

As shown in the below image, you have to scrape the above attributes.


```python
driver=webdriver.Chrome()
```

```python
#Opening the flipkart.com page on automated Chrome browser
driver.get("https://www.flipkart.com/")
```

```python
# Entering sneakers required in the question.

sneakers=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
sneakers.send_keys('sneakers')

search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
search.click()
```

```python
brand_name=[]
product=[]
price=[]
```

```python
#Scraping brand, product description & price from 3 pages
start=0
end=3

for page in range (start,end):
    tag_brand=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in tag_brand:
        brand_name.append(i.text)
           
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(3)
```

```python
start=0
end=3

for page in range (start,end):
    tag_product=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in tag_product:
        product.append(i.text)
           
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(3)
```

```python
start=0
end=3

for page in range (start,end):
    tag_price=driver.find_elements(By.XPATH,'//div[@class="_25b18c"]')
    for i in tag_price:
        price.append(i.text)
           
    next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(3)
```

```python
print(len(brand_name),len(product),len(price))
```

```python
brand=brand_name[:100]
description=product[:100]
cost=price[:100]
```

```python
df5=pd.DataFrame({'Brand Name':brand,'Product Description':description,'Price':cost})
```

# Q7: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” as shown in the below image:

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap 
a) Quote 
b) Author 
c) Type Of Quotes

```python
#Opening the azquotes.com page on automated Chrome browser

driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
top_quote=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quote.click()
```

```python
# Scrapting Quote, Author & Type of Quotes of top 1000 Quotes

quote=[]
quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote_tag:
    quote.append(i.text)
   
    
author=[]
author_tag=driver.find_elements(By.XPATH,'//span[@class="a-size-base s-underline-text"]')
for i in author_tag:
    author.append(i.text)
    
    
typeofquote=[]
typeofquote_tag=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in typeofquote_tag:
    typeofquote.append(i.text)
```

```python
start=0
end=10
quote=[]
author=[]
typeofquote=[]

for page in range (start,end):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag:
        quote.append(i.text)
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag:
        author.append(i.text)        
    typeofquote_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in typeofquote_tag:
        typeofquote.append(i.text)        
        
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
print(len(quote),len(author),len(typeofquote))
```

```python
df7=pd.DataFrame({'Quote':quote,'Author':author,'Type of Quote':typeofquote})
```

```python
df7
```

# Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
    
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ’50 most expensive cars’
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
#Opening the motor1.com page on automated Chrome browser

driver=webdriver.Chrome()
driver.get("https://www.motor1.com/")
```

```python
#searching 50 most expensive cars on motor1.com website.

car=driver.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input')
car.send_keys('50 most expensive cars')

search=driver.find_element(By.XPATH,'/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/button[1]')
search.click()
```

```python
searchcar=driver.find_element(By.XPATH,'/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/a/picture/img')
searchcar.click()
```

```python
# Scraping Car name & price of 50 most expensive cars.

car_name=[]
name_tag=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tag:
    car_name.append(i.text)
   
```

```python
name=car_name[:50]
```

```python

price=[]
price_tag=driver.find_elements(By.XPATH,'//strong[/html/body/div[10]/div[7]/div[2]/div[1]/div[2]/div[2]/p[4]/strong]')
for i in price_tag:
    price.append(i.text)
```

```python
print(len(name),len(price))
```

```python
df9=pd.DataFrame({'Car Name':name, 'Price':price})
```
