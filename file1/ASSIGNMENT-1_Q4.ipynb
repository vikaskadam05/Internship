```python
from bs4 import BeautifulSoup as bs
import requests
```

# 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-
a) Top 10 ODI teams in womenâ€™s cricket along with the records for matches, points and rating.


```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
```

```python
page
```

```python
soup=bs(page.content)
```

```python
soup
```

```python
# Top 10 Women Teams

team10=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    team10.append(i.text.replace('\n',''))
```

```python
top_10_team=team10[:10]
```

```python
top_10_team
```

```python
# Matches

matche1=[]
for i in soup.find_all('td',class_="rankings-block__banner--matches"):
    matche1.append(i.text)
```

```python
matche1
```

```python
# Matches

matches=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text)
```

```python
match=matches[0:18:2]
```

```python
match
```

```python
total_match=matche1+match
```

```python
total_match
```

```python
# Points

points1=[]
for i in soup.find_all('td',class_="rankings-block__banner--points"):
    points1.append(i.text)
```

```python
points1
```

```python
points=matches[1:19:2]
```

```python
points
```

```python
tota_points=points1+points
```

```python
tota_points
```

```python
# Rating

rating1=[]
for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating1.append(i.text.replace('\n','').replace('                            ','').replace('                            ',''))
```

```python
rating1
```

```python
N=9
rating9=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    rating9.append(i.text)
rat9=rating9[:N]
```

```python
rat9
```

```python
total_rating=rating1+rat9
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'Top Team':top_10_team, 'Matches':total_match,'Points':tota_points,'Rating':total_rating})
```

```python
df1
```
